```python
sleep_data=pd.read_csv("../data/raw/sleepDay_merged.csv")
```

```python
sleep_data['SleepDay']=pd.to_datetime(sleep_data['SleepDay']).dt.normalize()
```

# check for duplicates

```python
sleep_data.duplicated(subset=['Id','SleepDay']).sum()
```

```python
sleep_data.drop_duplicates(subset=['Id','SleepDay'])
```

```python
sleep_data.info()
sleep_data.describe()
```

# Save cleaned file

```python
sleep_data.to_csv("../data/cleaned/sleep_data_cleaned.csv",index=False)
```

```python
hourly_steps=pd.read_csv("../data/raw/hourlySteps_merged.csv")
```
